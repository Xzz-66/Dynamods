```python
# Imports
import numpy as np

# Probability matices
# Emmision probabilities [a u g c] [E I].T
emission = np.array([[0.25, 0.25, 0.25, 0.25],
                     [0.4, 0.4, 0.05, 0.15]])

# Transition probabilities [E I] [E, I].T
transition = np.array([[0.9, 0.1],
                       [0.2, 0.8]])

# Initial state probablities [E I]
initial = [0.5, 0.5]

# 'agcgc'
patient_alpha = [0,2,3,2,3]
# 'auuau'
patient_beta = [0,1,1,0,1]

def recursion(previous_delta, transition, emission, observation):
    
    psi = []
    delta = []
    for s in range(2):
        
        lst = [previous_delta[r] * transition[r, s] * emission[s, observation] for r in range(2)]
        
        arg_max = np.argmax(lst)
        max = np.max(arg_max)
        
        psi.append(arg_max)
        delta.append(max)
        
    return delta, psi 

def init_delta(initial, emission, observation):
    return [initial[s] * emission[s, observation] for s in range(2)]

def termination(delta, psi):

    final_state = psi[np.argmax(delta)]
    final_state_path = np.array(psi)[:,final_state]
    
    return final_state_path

def viterbi(initial, emission, transition, observation):
    
    # Initialization
    delta = init_delta(initial, emission, observation[0])
    
    # Recursion
    backtrack = []
    for o in observation[1:]:
        delta, psi = recursion(delta, transition, emission, o)
        backtrack.append(psi)
        
    # Termination
    final_state_path = termination(delta, backtrack)
    
    return final_state_path

alpha_state_path = viterbi(initial, emission, transition, patient_alpha)
beta_state_path = viterbi(initial, emission, transition, patient_beta)

# print(alpha_state_path)
# print(beta_state_path)
```
